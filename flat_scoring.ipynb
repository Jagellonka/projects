```python
leaving_area_mean = data['living_area'].mean() / data['total_area'].mean() # —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏ –∫ –æ–±—â–µ–π
kitchen_area_mean = data['kitchen_area'].mean() / data['total_area'].mean() # —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ –∫ –æ–±—â–µ–π
data['living_area'].fillna((data['total_area'] * leaving_area_mean).round(2), inplace=True) # –∑–∞–ø–æ–ª–Ω—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏
data['kitchen_area'].fillna((data['total_area'] * kitchen_area_mean).round(2), inplace=True) # –∑–∞–ø–æ–ª–Ω—è–µ–º –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏
data[['living_area', 'kitchen_area']].isna().sum() # –ø—Ä–æ–≤–µ—Ä–∫–∞
```

```python
data['locality_name'] = data['locality_name'].str.lower() # —É–º–µ–Ω—å—à–∞–µ–º —Ä–µ–≥–∏—Å—Ç—Ä –Ω–∞–∑–≤–∞–Ω–∏–π
data['locality_name'] = data['locality_name'].str.replace('—ë', '–µ') # –º–µ–Ω—è–µ–º —ë –Ω–∞ –µ
data['locality_name']
```

–°–æ—Å—Ç–∞–≤–∏–º —Å–ª–æ–≤–∞—Ä—å –∏–∑ —Å–ª–æ–≤ –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö –ª–æ–∫–∞—Ü–∏–π –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ.

```python
locality_name_list = data['locality_name'].to_list() # —Å–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –∏–∑ —Ç–æ–ø–æ–Ω–∏–º–æ–≤
locality_name_dict = {} # –æ–±—ä—è–≤–ª—è–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Å–ª–æ–≤ –∏–∑ —Ç–æ–ø–æ–Ω–∏–º–æ–≤
for locality_name in locality_name_list: # —Å–æ–∑–¥–∞–µ–º —Ü–∏–∫–ª, –≤ –∫–æ—Ç–æ—Ä–æ–º –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å–ø–∏—Å–∫–∞ –∏–∑ —Ç–æ–ø–æ–Ω–∏–º–æ–≤
    splitted_name = locality_name.split() # —Ä–∞–∑–±–∏–≤–∞–µ–º –∫–∞–∂–¥–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞
    for name in splitted_name: # —Å–æ–∑–¥–∞–µ–º –ø–æ–¥—Ü–∏–∫–ª, –≤ –∫–æ—Ç–æ—Ä–æ–º –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ª–æ–≤–∞–º –∏–∑ –∫–∞–∂–¥–æ–≥–æ –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞
        locality_name_dict[name] = locality_name_dict.get(name, 0) + 1 # –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ —Å–ª–æ–≤–∞—Ä—å —Å–ª–æ–≤–∞ –∏ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
locality_name_dict
```

–°–ª–æ–≤–∞—Ä—å –≤—ã—à–µ –Ω—É–∂–µ–Ω, —á—Ç–æ–±—ã —É–¥–æ–±–Ω–æ –±—ã–ª–æ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –∫–∞–∫–∏–µ —Å–ª–æ–≤–∞ –≤–æ–æ–±—â–µ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö. –£–¥–∞–ª–∏–º –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏–π —Å–ª–æ–≤–∞ –ø–æ —Ç–∏–ø—É: '–ø–æ—Å–µ–ª–æ–∫', '–≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ' –∏ —Ç.–¥

```python
# —Å–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –∏–∑ —Å–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —É–¥–∞–ª–∏–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏
undue_words = ['–ø–æ—Å–µ–ª–æ–∫ ', '–≥–æ—Ä–æ–¥—Å–∫–æ–π ', '–¥–µ—Ä–µ–≤–Ω—è ', '–≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ ', '—Ç–∏–ø–∞ ', '—Å–∞–¥–æ–≤–æ–µ ', '—Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ ', '—Å–µ–ª–æ ']
for word in undue_words: # —Å–æ–∑–¥–∞–µ–º —Ü–∏–∫–ª –∏ –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ –∫–∞–∂–¥–æ–º—É —Å–ª–æ–≤—É –∏–∑ —Å–ø–∏—Å–∫–∞ –≤—ã—à–µ
    data['locality_name'] = data['locality_name'].str.replace(word, '') # —É–¥–∞–ª—è–µ–º —Å–ª–æ–≤–æ –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏–π –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤
data['locality_name'].nunique() # —Å–º–æ—Ç—Ä–∏–º —Å–∫–æ–ª—å–∫–æ –æ—Å—Ç–∞–ª–æ—Å—å —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π 
```

```python
def compare(data):
    q1 = data.quantile(q=.25) # –Ω–∞—Ö–æ–¥–∏–º –ø–µ—Ä–≤—ã–π –∫–≤–∞—Ä—Ç–∏–ª—å
    q3 = data.quantile(q=.75) # –Ω–∞—Ö–æ–¥–∏–º —Ç—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å
    iqr = q3 - q1 # —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º –∏ —Ç—Ä–µ—Ç—å–∏–º –∫–≤–∞—Ä—Ç–∏–ª–µ–º
    return q3+1.5*iqr # –Ω–∞—Ö–æ–¥–∏–º –≤–µ—Ä—Ö–Ω–∏–π –¥–∏–∞–ø–∞–∑–æ–Ω —Ä–∞–∑–º–∞—Ö–∞
total_area_compare = compare(data['total_area'])
# —Å—Ç—Ä–æ–∏–º –∫–æ—Ä–æ–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –¥–ª—è –±–æ–ª—å—à–∏—Ö –∫–≤–∞—Ä—Ç–∏—Ä
data.query('total_area > @total_area_compare').boxplot(column='total_area', figsize=(9,12));
```
